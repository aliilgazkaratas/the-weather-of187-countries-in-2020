```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import statsmodels.api as sm
```

```python
dataset = pd.read_csv("the weather of 187 countries in 2020.csv")
```

```python
dataset.tail()
```

```python
dataset["TAVG_ATTRIBUTES"].isna().sum()
```

```python
#So there is a suspicious about columns may have lots of NaN value, 
#we have to find those columns and remove them.
```

```python
print("*****Percentage of NaN values*****")
for i in range(0,len(dataset.columns.values)):
    print(dataset.columns.values[i]," ==> %",(dataset[dataset.columns.values[i]].isna().sum()/dataset.shape[0])*100)
```

```python
# I will hold PRCP, TAVG, TMAX ,and TMIN columns for now
```

```python
print("*****Percentage of NaN values*****")
list_of_holdings=[]
for i in range(0,len(dataset.columns.values)):
    percentage=((dataset[dataset.columns.values[i]].isna().sum())/(dataset.shape[0]))*100
    print(dataset.columns.values[i]," ==> %",percentage)
    if percentage<60:
        list_of_holdings.append(dataset.columns.values[i])
    else:
        continue
print("List that only has known ones: ",list_of_holdings)
        
```

```python
dataset[list_of_holdings]
```

```python
# Number of Countries in our Dataset
grouped_country_dataset = dataset["Country/Region"].nunique()
print("Number of Countries:",grouped_country_dataset)
```

```python
# Group the dataset by "Country/Region" to see average Temperatures for each one.
meanTemperature_country_dataset = dataset.groupby("Country/Region")["TAVG"].mean().sort_values(ascending=False)
print("Mean Temperature of Countries:",meanTemperature_country_dataset)
```

```python
#The most hot ones
meanTemperature_country_dataset.head(10)
```

```python
#The most cold ones
meanTemperature_country_dataset.tail(10).sort_values(ascending=True)
```

#Let's find that whether a linear regression with average temperature and precipitation

```python
# Group the dataset by "Country/Region" to see average Precipitation for each one.
meanPrecipitation_country_dataset = dataset.groupby("Country/Region")["PRCP"].mean().sort_values(ascending=False)
print("Mean Precipitation of Countries:",meanPrecipitation_country_dataset)
```

```python
# Assuming you have two datasets: dataset1 and dataset2

# Concatenate the datasets vertically
concatenated_dataset = pd.concat([meanTemperature_country_dataset, meanPrecipitation_country_dataset], axis=1)

# Verify the concatenated dataset
print(concatenated_dataset)
# Assuming you have a dataset named 'concatenated_dataset'

# Remove rows with NaN values
concatenated_dataset = concatenated_dataset.dropna()

# Verify the updated dataset
print(concatenated_dataset)
```

```python
# Define the independent variable (X) and dependent variable (y)
X = concatenated_dataset.iloc[:, -2]  # Replace '-2' with the appropriate column index
y = concatenated_dataset.iloc[:, -1]  # Replace '-1' with the appropriate column index

# Create the linear regression model
model = sm.OLS(y, X)

# Add a constant to the independent variable
X = sm.add_constant(X)

# Fit the model to the data
results = model.fit()

# Print the regression results
print(results.summary())
```

```python
import matplotlib.pyplot as plt

# Assuming you have a dataset named 'meanTemperature_country_dataset'

# Group the dataset by country and calculate the average temperature
mean_temperatures = dataset.groupby('Country/Region')['TAVG'].mean()

# Sort the mean temperatures in descending order
mean_temperatures = mean_temperatures.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 10))
plt.bar(mean_temperatures.index, mean_temperatures)
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.title('Average Temperatures by Country')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the chart
plt.show()
```
